```python
def read_file(tenfile):
    f = open(tenfile,encoding="utf8")
    data = f.readlines()
    header = data[0].strip().split(",")
    D_data = {}
    for i in header:
        D_data[i] = []
    data = data[1:]
    for i in data:
        i = i.strip().split(",")
        for j in range(len(header)):
            D_data[header[j]].append(i[j])
    return D_data

def insert_file(tenfile,data):
    f = open(tenfile,"a",encoding='utf8')
    f.write(data)
    f.close()

from tkinter import *

sc = Tk()
sc.title("Login screen")
sc.geometry("450x200")

Data_sp = read_file("sanpham.csv")

def them():
    Sp = var_et_them.get()
    Gia = var_et_gia.get()
    DV = var_et_dv.get()
    Data_sp["Tên sản phẩm"].append(Sp)
    Data_sp["Giá"].append(Gia)
    Data_sp["Đơn vị"].append(DV)
    insert_file("sanpham.csv",Sp+","+Gia+","+DV+"\n")
    var_et_them.set("")
    var_et_gia.set("")
    var_et_dv.set("")

def timkiem():
    Sp = var_et_tk.get()
    if Sp not in Data_sp["Tên sản phẩm"]:
        lb_tb.config(text = "Chưa có tên sản phẩm này")
        return
    else:
        index = Data_sp["Tên sản phẩm"].index(Sp)
        lb_tb.config(text = "")
        lb_tbgia.config(text=Data_sp["Giá"][index])
        lb_tbdv.config(text=Data_sp["Đơn vị"][index])
        return

var_et_them = StringVar()
var_et_gia = StringVar()
var_et_dv = StringVar()
var_et_tk = StringVar()

lb_them = Label(sc,text = "THÊM")
lb_them.grid(column=2,row=1)

lb_sp= Label(sc,text = "Tên sản phẩm")
lb_sp.grid(column=1,row=2)

et_sp = Entry(sc,textvariable = var_et_them)
et_sp.grid(column=2,row=2)

lb_gia= Label(sc,text = "Giá")
lb_gia.grid(column=1,row=3)

et_gia = Entry(sc,textvariable = var_et_gia)
et_gia.grid(column=2,row=3)

lb_dv= Label(sc,text = "Đơn vị")
lb_dv.grid(column=1,row=4)

et_dv = Entry(sc,textvariable = var_et_dv)
et_dv.grid(column=2,row=4)

bt_them = Button(sc,text="THÊM",command=them)
bt_them.grid(column=2,row=5)

#Tìm Kiếm

lb_tk = Label(sc,text = "TÌM KIẾM",padx=40,pady=20)
lb_tk.grid(column=4,row=1)

lb_tsp = Label(sc,text = "Tên sản phẩm",padx=10,pady=5)
lb_tsp.grid(column=3,row=2)

et_tk = Entry(sc,textvariable = var_et_tk)
et_tk.grid(column=4,row=2)

bt_tk = Button(sc,text="Tìm",command = timkiem)
bt_tk.grid(column=4,row=3)

lb_giaht = Label(sc,text = "Giá")
lb_giaht.grid(column=3,row=4)

lb_tbgia = Label(sc,text="")
lb_tbgia.grid(column=4,row=4)

lb_dvht = Label(sc,text = "Đơn vị")
lb_dvht.grid(column=3,row=5)

lb_tbdv = Label(sc,text="")
lb_tbdv.grid(column=4,row=5)

lb_tb = Label(sc,text="")
lb_tb.grid(column=4,row=6)

sc.mainloop()
```
