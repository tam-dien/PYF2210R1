### Thêm widget cho cửa sổ

Widget là những đối tượng sẽ hiển thị và hoạt động trên cửa số

Có rất nhiều loại Widget như: nút nhấn (button), nhãn (label), khung nhập dữ liệu (Entry),...

### Label

Cú pháp:

`<tên đối tượng> = Label(<cửa số>,text = <nội dung>)`

##### Lưu ý, ta cần dùng lệnh `<tên đối tượng>.pack()` để Label có thể hiển thị lên cửa sổ

<b> Hãy thêm một đối tượng Label vào cửa sổ: </b>

```python
from tkinter import *
sc = Tk() # chỉ được tạo 1 cửa sổ cho 1 chương trình 
sc.title("hello Python")
lb1 = Label(sc,text = "Tkinter") #có thể thay đổi màu nền, màu chữ, font chữ, size chữ để giao diện dễ nhìn hơn
lb1.pack()
lb2 = Label(sc,text = "Class")
lb2.pack()
sc.mainloop() #~> lệnh để cửa sở giao diện hoạt động
```

### Entry

Cú pháp:

`<tên đối tượng> = Entry(<cửa số>)`

##### Lưu ý, ta cần dùng lệnh `<tên đối tượng>.pack()` để Entry có thể hiển thị lên cửa sổ

<b> Hãy thêm một đối tượng Entry vào cửa sổ: </b>

```python
from tkinter import *
sc = Tk() # chỉ được tạo 1 cửa sổ cho 1 chương trình 
sc.title("hello Python")
lb1 = Label(sc,text = "Tkinter") #có thể thay đổi màu nền, màu chữ, font chữ, size chữ để giao diện dễ nhìn hơn
lb1.pack()
lb2 = Label(sc,text = "Class")
lb2.pack()
def click(): # vị trí đặt hàm phải để trước sự kiện(bấm nút, điền thông tin)
    print(" tôi đang học Python")
bt1 = Button(sc,text="List",command=click) # tạo nút bấm, lệnh command là để liên kết câu sự kiện bấm nút với hàm thao tác 
bt1.pack()
et1 = Entry(sc) #tạo bảng điền thông tin
et1.pack()
sc.mainloop() #~> lệnh để cửa sở giao diện hoạt động
```

### pack()

Lệnh `pack()` sẽ giúp sắp xếp các đối tượng theo thứ tự nhất định

pack() có các tham số:

- <b>expand</b>:  Tùy chọn này sẽ quyết định các widget có được mở rộng để lấp đầy bất kì khoảng trống thừa trong tống thể hình học hay không. Nó có các giá trị Boolean, True hoặc False, giá trị mặc định là False
- <b>fill</b>: Nó được sử dụng để lấp đầy các khoảng trống thừa . Giá trị mặc định là NONE, và có giá định khác như X (lấp theo chiều ngang) , Y(lấp theo chiều dọc) hoặc BOTH (lấp theo chiều ngang và dọc)
- <b>side</b>: Được sử dụng để chỉ hướng mặt của widget được đóng gói bên trong cửa sổ , có các giá trị TOP(mặc định) ,BOTTOM, LEFT hoặc RIGHT

<b>Hãy ứng dụng lệnh pack() để xây dựng cửa sổ sau:</b>



```python
from tkinter import *
sc = Tk()
sc.title("login screen")
sc.geometry("350x200")
### khi click vào login, nếu người dùng nhập username là 'python' và password là 'cơ bản' thì thông báo login thành công
### ngược lại thông báo login thất bại
### thông báo sẽ được hiển thị trên lb_thongbao, nếu login thành công thì thông báo chữ màu xanh
### nếu login thất bại, thông báo chữ màu đỏ
def login():
    if var_et1.get()=="python" and var_et2.get()=="cơ bản":
        lb_thongbao.config(text="Login thành công!",fg="green")
    else:
        lb_thongbao.config(text="Login thất bại!",fg="red")
        
var_et1 = StringVar()
var_et2 = StringVar()

lb1 = Label(sc,text="Username",fg="black",font=("Arial",),width=20)
lb1.pack()
et1 = Entry(sc,fg="black",font=("Arial",),width=20,textvariable=var_et1)
et1.pack()
lb2 = Label(sc,text="Password",fg="black",font=("Arial",),width=20)
lb2.pack()
et2 = Entry(sc,fg="black",font=("Arial",),width=20,textvariable=var_et2)
et2.pack()

bt1 = Button(sc,text="Login",fg="black",font=("Arial",),width=5,command=login)
bt1.pack()

lb_thongbao = Label(sc,text="",font=("Arial",),width=20)
lb_thongbao.pack()

sc.mainloop()
```

### grid()

Lệnh `grid()` sẽ giúp sắp xếp các đối tượng theo một bảng

grid() có các tham số:

- column : Tùy chọn này chỉ định số cột mà widget sẽ được đặt. Giá trị mặc định là 0
- columnspan:  Dùng chỉ định số lượng cột mà widget sẽ được mở rộng
- row: Tùy chọn này chỉ định số hàng cho widget sẽ được đặt .  Giá trị mặc định là 0
- rowspan: Dùng chỉ định số lượng hàng mà widget sẽ được mở rộng
- padx, pady: Dùng thể hiện pixel của padding được thêm vào bên ngoài đường viền widget . padx,  là phần padding theo chiều ngang, còn pady là phần padding theo chiều dọc.
- ipadx, ipady : Dùng thể hiện pixel của padding được thêm vào bên trong đường viền widget. ipadx,  là phần padding theo chiều ngang, còn ipady là phần padding theo chiều dọc.
- sticky : Được sử dụng trong điều kiện ô lớn hơn widget, được sử dụng chỉ định vị trí của widget bên trong ô. Nếu không sử dụng sticky, thì widget mặc định đặt ở giữa ô. Sticky có thể có các giá trị như N, E, S, W, NE, NW, SE & SW.

<b>Hãy ứng dụng lệnh grid() để xây dựng cửa sổ sau:</b>



```python
# chỉ được dùng gird hoặc pack
from tkinter import *
sc = Tk()
sc.geometry("350x200")
def login():
    if if var_et1.get()=="python" and var_et2.get()=="cơ bản":
        lb_thongbao.config(text="Login thành công!",fg="green")
    else:
        lb_thongbao.config(text="Login thất bại!",fg="red")
var_et1 = StringVar()
var_et2 = StringVar()
lb1 = Label(sc,text="Username",fg="black",font=("Arial"))
lb1.grid(column = 1, row = 2 )
lb2 = Label(sc,text="Password",fg="black",font=("Arial"))
lb2.grid(column = 1, row = 3 )
lb3 = Label(sc,text="Please enter email and password to login",fg="black",font=("Arial"))
lb3.grid(column = 2, row = 1 )
et1 = Entry(sc,fg="black",font=("Arial",),width=20,textvariable=var_et1)
et1.grid(column = 2, row = 2 )
et2 = Entry(sc,fg="black",font=("Arial",),width=20,textvariable=var_et2)
et2.grid(column = 2, row = 3 )
bt1 = Button(sc,text="Login",fg="black",font=("Arial",),width=5,command=login)
bt1.grid(column = 2, row = 4 )
lb_thongbao = Label(sc,text="",font=("Arial",),width=20)
lb_thongbao.grid(column = 2, row = 5 )
sc.mainloop()
```

```python
def update():
    fr_update.pack()
    fr_main.pack_forget()
def back():
    fr_main.pack()
    fr_update.pack_forget() or fr_find.pack_forget()
    var_et1_find.set("")
    var_et2_find.set("")
    var_et3_find.set("")
    var_et1_update.set("")
    var_et2_update.set("")
    var_et3_update.set("")
def find():
    fr_find.pack()
    fr_main.pack_forget()
def readfile(tenfiles): #đọc file
    f = open(tenfiles)
    data = f.readlines()
    header = data[0].strip().split(",")
    D_data = {}
    for i in header:
        D_data[i] = []
    data= data[1:]
    for i in data:
        i = i.strip().split(",")
        for j in range(len(i)):
            D_data[header[j]].append(i[j])
    return D_data

def insert_file(tenfile,data): #thêm dữ liệu
    f = open(tenfile,"a",encoding='utf8')
    f.write(data)
    f.close()

def tim_sp(): # tìm sản phẩm
    if var_et1_find.get() not in D_data['name']:
        lb_tb_find.config(text="sản phẩm không tồn tại!",fg="red")
        return
    else:
        index = D_data['name'].index(var_et1_find.get())
        var_et2_find.set(D_data['price'][index])
        var_et3_find.set(D_data['unit'][index])
        return
def them_sp(): # thêm sản phẩm
    name = var_et1_update.get()
    price = var_et2_update.get()
    unit = var_et3_update.get()
    if name in D_data['name'] or name =="" or price == "" or unit == "":
        lb_tb_update.config(text="Thông tin nhập không hợp lệ!",fg="red")
        return
    else:
        D_data['name'].append(name)
        D_data['price'].append(price)
        D_data['unit'].append(unit)
        insert_file("sp.csv",name+","+price+","+unit+"\n")
        lb_tb_update.config(text="Thêm sản phẩm thành công",fg="green")
        var_et1_update.set("")
        var_et2_update.set("")
        var_et3_update.set("")
from tkinter import *
sc = Tk()
sc.geometry("550x300")
sc.title("sản phẩm")
D_data = readfile("sp.csv")
#khởi tạo các giao diện sử dụng
fr_main = Frame(sc)
fr_update = Frame(sc)
fr_find = Frame(sc)
#màn hình main
lb_main = Label(fr_main,text="Quản lý sản phẩm",fg="black",font=("Arial"))
lb_main.grid(column = 1, row = 1 )
bt_update = Button(fr_main,text="Thêm sản phẩm",fg="black",font=("Arial",),width=25,command=update)
bt_update.grid(column = 1, row = 2 )
bt_find = Button(fr_main,text="Tìm kiếm sản phẩm",fg="black",font=("Arial",),width=25,command=find)
bt_find.grid(column = 1, row = 3 )
fr_main.pack()
#màn hình update
var_et1_update = StringVar()
var_et2_update = StringVar()
var_et3_update = StringVar()
lb_update = Label(fr_update,text="Thêm sản phẩm",fg="black",font=("Arial"),)
lb_update.grid(column = 1, row = 1,columnspan = 2 )
lb1_update = Label(fr_update,text="Tên sản phẩm",fg="black",font=("Arial"))
lb1_update.grid(column = 1, row = 2 )
lb2_update = Label(fr_update,text="Giá",fg="black",font=("Arial"))
lb2_update.grid(column = 1, row = 3 )
lb3_update = Label(fr_update,text="Đơn vị tính",fg="black",font=("Arial"))
lb3_update.grid(column = 1, row = 4 )
lb_tb_update = Label(fr_update,text="",font=("Arial",),width=30)
lb_tb_update.grid(column = 1, row = 7, columnspan = 2  )
et1_update = Entry(fr_update,fg="black",font=("Arial",),width=20,textvariable=var_et1_update)
et1_update.grid(column = 2, row = 2 )
et2_update = Entry(fr_update,fg="black",font=("Arial",),width=20,textvariable=var_et2_update)
et2_update.grid(column = 2, row = 3 )
et3_update = Entry(fr_update,fg="black",font=("Arial",),width=20,textvariable=var_et3_update)
et3_update.grid(column = 2, row = 4 )
bt1_update = Button(fr_update,text="Thêm",fg="black",font=("Arial",),width=15,command=them_sp)
bt1_update.grid(column = 1, row = 5, columnspan = 2 )
bt2_update = Button(fr_update,text="Trở lại",fg="black",font=("Arial",),width=15,command=back)
bt2_update.grid(column = 1, row = 6, columnspan = 2 )
#màn hình find:
var_et1_find = StringVar()
var_et2_find = StringVar()
var_et3_find = StringVar()
lb_find = Label(fr_find,text="Tìm kiếm sản phẩm",fg="black",font=("Arial"),)
lb_find.grid(column = 1, row = 1,columnspan = 2 )
lb1_find = Label(fr_find,text="Tên sản phẩm",fg="black",font=("Arial"))
lb1_find.grid(column = 1, row = 2 )
lb2_find = Label(fr_find,text="Giá",fg="black",font=("Arial"))
lb2_find.grid(column = 1, row = 4 )
lb3_find = Label(fr_find,text="Đơn vị tính",fg="black",font=("Arial"))
lb3_find.grid(column = 1, row = 5)
lb_tb_find = Label(fr_find,text="",font=("Arial",),width=20)
lb_tb_find.grid(column = 1, row = 7, columnspan = 2  )
et1_find = Entry(fr_find,fg="black",font=("Arial",),width=20,textvariable=var_et1_find)
et1_find.grid(column = 2, row = 2 )
et2_find = Entry(fr_find,fg="black",font=("Arial",),width=20,textvariable=var_et2_find)
et2_find.grid(column = 2, row = 4 )
et3_find = Entry(fr_find,fg="black",font=("Arial",),width=20,textvariable=var_et3_find)
et3_find.grid(column = 2, row = 5 )
bt1_find = Button(fr_find,text="Tìm",fg="black",font=("Arial",),width=15,command=tim_sp)
bt1_find.grid(column = 1, row = 3, columnspan = 2 )
bt2_find = Button(fr_find,text="Trở lại",fg="black",font=("Arial",),width=15,command=back)
bt2_find.grid(column = 1, row = 6, columnspan = 2  )
sc.mainloop()
```
