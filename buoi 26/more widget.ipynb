```python
### Radio Button

from tkinter import *

sc = Tk()

sc.geometry("500x300")

lb1 = Label(sc,text="Chọn đơn vị")
lb1.pack()

def click_me():
    print(rb_var.get())
    
def click_radio():
    print("Bạn đã click và radio button và giá trị của radio là:",rb_var.get())

rb_var = IntVar()
### khi vừa khởi chạy rb_var sẽ có giá trị là 0
rb_var.set(1)

rb1 = Radiobutton(sc,text="Cái",variable=rb_var,value=1,command=click_radio)
rb1.pack()

rb2 = Radiobutton(sc,text="Kg",variable=rb_var,value=2,command=click_radio)
rb2.pack()

rb3 = Radiobutton(sc,text="Dĩa",variable=rb_var,value=3,command=click_radio)
rb3.pack()

bt1 = Button(sc,text="bấm tui đi",command=click_me)
bt1.pack()

sc.mainloop()
```

```python
## checkbox

from tkinter import *

sc = Tk()

sc.geometry("500x300")

lb1 = Label(sc,text="Chọn nhóm hàng")
lb1.pack()

L_loaihang = [] ### nó sẽ lưu danh sách loại hàng

### giả sử ta click hết 3 cái checkbox thì list L_loaihang = ["hàng gia dụng","hàng nội thất","hàng ngoại thất"]

def click_check1():
    if cb1_var.get() ==1:
        if "Hàng gia dụng" not in L_loaihang:
            L_loaihang.append("Hàng gia dụng")
    else:
        if "Hàng gia dụng" in L_loaihang:
            L_loaihang.remove("Hàng gia dụng")
    print(L_loaihang)
    return L_loaihang
    
def click_check2():
    if cb2_var.get() ==1:
        if "Hàng nội thất" not in L_loaihang:
            L_loaihang.append("Hàng nội thất")
    else:
        if "Hàng nội thất" in L_loaihang:
            L_loaihang.remove("Hàng nội thất")
    print(L_loaihang)
    
def click_check3():
    if cb3_var.get() ==1:
        if "Hàng ngoại thất" not in L_loaihang:
            L_loaihang.append("Hàng ngoại thất")
    else:
        if "Hàng ngoại thất" in L_loaihang:
            L_loaihang.remove("Hàng ngoại thất")
    print(L_loaihang)
    return L_loaihang

cb1_var = IntVar()

cb1 = Checkbutton(sc,text="Hàng gia dụng",variable=cb1_var,onvalue=1,offvalue=0,command=click_check1)
cb1.pack()

cb2_var = IntVar()

cb2 = Checkbutton(sc,text="Hàng nội thất",variable=cb2_var,onvalue=1,offvalue=0,command=click_check2)
cb2.pack()

cb3_var = IntVar()

cb3 = Checkbutton(sc,text="Hàng ngoại thất",variable=cb3_var,onvalue=1,offvalue=0,command=click_check3)
cb3.pack()

sc.mainloop()
```

```python
from tkinter import *

sc = Tk()

sc.geometry("500x300")

lb1 = Label(sc,text="Chọn nhóm hàng")
lb1.pack()

L_loaihang = []

### 3 cái checkbox đều có 1 biến là Checkbutton, 1 biến là IntVar và 1 hàm để bắt sự kiện click vào checkbutton

class checkbox_nhomhang:
    def __init__(self,text):
        self.cb_var = IntVar()
        self.text = text
        self.cb = Checkbutton(sc,text=text,variable=self.cb_var,onvalue=1,offvalue=0,command=self.click_check)
        self.cb.pack()
    def click_check(self):
        if self.cb_var.get()==1:
            if self.text not in L_loaihang:
                L_loaihang.append(self.text)
        else:
            if self.text in L_loaihang:
                L_loaihang.remove(self.text)
        print(L_loaihang)
        
checkbox_1 = checkbox_nhomhang("Hàng nội thất")
checkbox_2 = checkbox_nhomhang("Hàng gia dụng")
checkbox_3 = checkbox_nhomhang("Hàng ngoại thất")

sc.mainloop()
```

```python
## combobox (select list)

from tkinter import *
from tkinter import ttk

sc = Tk()

sc.geometry("500x300")

lb1 = Label(sc,text="Chọn giới tính")
lb1.pack()

cbb1_var = StringVar()

cbb1 = ttk.Combobox(sc,textvariable=cbb1_var,state="readonly")
cbb1.pack()

cbb1['values'] = ["Nam","Nữ","Không xác định"]

def click_me():
    print(cbb1_var.get())

bt1 = Button(sc,text="bấm tui đi",command=click_me)
bt1.pack()

sc.mainloop()
```
